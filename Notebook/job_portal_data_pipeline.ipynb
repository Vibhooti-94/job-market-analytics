```python
query = """
WITH company_jobs AS (
    SELECT
    employer_id,
    COUNT(*) AS total_jobs
    FROM jobs
    GROUP BY employer_id
)

SELECT
e.company_name,
cj.total_jobs
FROM company_jobs cj
JOIN employers e
ON cj.employer_id = e.employer_id
ORDER BY total_jobs DESC
LIMIT 10
"""

cursor.execute(query)

print(cursor.fetchall())
```

Create Trigger (Automation)

Example: Automatically log new applications

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS application_logs (
    log_id SERIAL PRIMARY KEY,
    application_id INT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
""")

conn.commit()
```

```python
cursor.execute("""
CREATE OR REPLACE FUNCTION log_new_application()
RETURNS TRIGGER AS $$
BEGIN
INSERT INTO application_logs(application_id)
VALUES (NEW.application_id);

RETURN NEW;
END;
$$ LANGUAGE plpgsql;
""")
```

```python
cursor.execute("""
CREATE TRIGGER application_insert_trigger
AFTER INSERT ON applications
FOR EACH ROW
EXECUTE FUNCTION log_new_application();
""")

conn.commit()
```

Transaction Management in Google Colab
Transaction Flow:

Start Transaction
      ↓
Execute Queries
      ↓
Commit  (save changes permanently)
or
Rollback (undo changes)

```python
##COMMIT saves the changes permanently in the database.Example: Insert a new job and commit.
```

```python
cursor.execute("""
INSERT INTO jobs (employer_id,job_title,location,salary_min,salary_max)
VALUES (2,'Junior Data Analyst','Hyderabad',500000,800000)
""")

conn.commit()

print("Transaction committed successfully")
```

```python
##ROLLBACK (Undo Transaction)-If something goes wrong, we undo changes using rollback.
```

```python
try:

    cursor.execute("""
    INSERT INTO skills(skill_name)
    VALUES ('Big Data')
    """)

    cursor.execute("""
    INSERT INTO skills(skill_name)
    VALUES ('Big Data')
    """)

    conn.commit()

except Exception as e:

    conn.rollback()

    print("Error occurred, transaction rolled back")
```

Explanation:

First insert → success
Second insert → error (duplicate)
Rollback → both inserts cancelled

This prevents bad data from entering the database.

```python
##SAVEPOINT (Partial Rollback)-Sometimes we want to rollback only part of the transaction.
```

```python
cursor.execute("BEGIN")

cursor.execute("""
INSERT INTO skills(skill_name)
VALUES ('Cloud Computing')
""")

cursor.execute("SAVEPOINT before_error")

try:

    cursor.execute("""
    INSERT INTO skills(skill_name)
    VALUES ('Cloud Computing')
    """)

except:

    cursor.execute("ROLLBACK TO SAVEPOINT before_error")

conn.commit()

print("Savepoint example completed")
```

Insert 1 → success
SAVEPOINT created
Insert 2 → error
Rollback to savepoint
First insert remains
Second insert cancelled

```python
try:

    cursor.execute("BEGIN")

    cursor.execute("""
    INSERT INTO applications(job_id,seeker_id,status)
    VALUES (10,20,'Applied')
    """)

    cursor.execute("""
    UPDATE jobs
    SET salary_max = salary_max + 10000
    WHERE job_id = 10
    """)

    conn.commit()

    print("Application transaction completed")

except:

    conn.rollback()

    print("Transaction failed and rolled back")
```
